```python
import pandas as pd
import numpy as np
from math import sqrt
from sklearn.model_selection import train_test_split, KFold, cross_val_predict, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
import time
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
matrices = np.load('/Users/XIAN/Documents/xian fyp/Coulomb Matrix/Coulomb Matrix 44.npy')
molden = np.load('/Users/XIAN/Documents/xian fyp/Coulomb Matrix/Molden 44 eV.npy')
print(matrices.shape)
print(molden.shape)
n_samples = len(matrices)
data = matrices.reshape((n_samples, -1))
print(data.shape)
```

```python
X = data
y = molden
X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.25, random_state=42)
```

```python
start = time.time()
mlp = MLPRegressor(activation= 'logistic', alpha=0.000005, hidden_layer_sizes=(60,54,48,42),max_iter=50000,learning_rate_init =0.001,learning_rate = 'constant')
mlp.fit(X_train,y_train)
predictions = mlp.predict(X_test)
score = mlp.score(X_test, y_test)
print(score)
end = time.time()
print(end - start)
from sklearn.metrics import mean_squared_error
MAE2 = mean_squared_error(y_test, predictions, multioutput='uniform_average')
print(MAE2)
print(sqrt(MAE2))
```

```python
kf = KFold(4, shuffle=True)
myscore = 0
for train_indices, test_indices in kf.split(X):
    mlp = MLPRegressor(activation= 'logistic', alpha=0.000005, hidden_layer_sizes=(60,54,48,42),max_iter=1000,learning_rate_init =0.001,learning_rate = 'adaptive')
    mlp.fit(X[train_indices], y[train_indices])
    print(mlp.score(X[test_indices], y[test_indices]))
    myscore += mlp.score(X[test_indices], y[test_indices])
print(myscore/4)
```

```python
activations = ['logistic']
solvers = ['adam']

learning_rates = ['constant']

max_iters = [400]
hidden_layer_sizess = [(30,27,24,21),(16,32,16,32)]
alphas = [0.000008,0.000005,0.000001,0.0000008]
learning_rate_inits = [0.05,0.01,0.005,0.001]

regressors = []
for a in activations:
    for b in solvers:
        for d in learning_rates:
            for g in hidden_layer_sizess:
                for c in alphas:
                    for e in learning_rate_inits:
                        for f in max_iters:
                            regressors.append(MLPRegressor(activation=a, solver=b, alpha=c,learning_rate=d, learning_rate_init=e,max_iter=f, hidden_layer_sizes=g, random_state=25))
```

```python
y_pred = cross_val_predict(mlp, X, y, cv=3)
print(cross_val_score(mlp, X, y, cv=4))

```
